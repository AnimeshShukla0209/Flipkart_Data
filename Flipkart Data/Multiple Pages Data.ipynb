```python
# Importing Important Libraries
from tqdm import tqdm
import pandas as pd
import numpy as np
from tqdm import tqdm
import requests
import time

from bs4 import BeautifulSoup
from selenium import webdriver
import time
from tqdm import tqdm

link = "https://www.flipkart.com/search?q=phones&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off&page=" + i
```

```python
Multiple_Pages = []
for i in tqdm(range(100)):
    link = "https://www.flipkart.com/search?q=phones&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off&page=" + str(i)
    options = webdriver.EdgeOptions()
    driver = webdriver.Edge(options=options)
    driver.get(link)
    soup = BeautifulSoup(driver.page_source, "html.parser")
    time.sleep(7)
    for sp2 in (soup.find_all("div",class_="_2kHMtA")):
        try:
            Model_Name = sp2.find_all("div",class_="_4rR01T")[0].text.split("(")[0][:-1]
        except:
            None
        try:
            Rating = sp2.find_all("span",class_="_1lRcqv")[0].text
        except:
            None
        try:    
            Ram = sp2.find_all("li",class_="rgWa7D")[0].text.split("|")[0][:-7]
        except:
            None
        try:
            ROM = sp2.find_all("li",class_="rgWa7D")[0].text.split("|")[1][:-7]
        except:    
            None
        try:    
            Screen_Size = sp2.find_all("li",class_="rgWa7D")[1].text[:5] + " cm"
        except:    
            None
        try:    
            Rear_Camera = sp2.find_all("li",class_="rgWa7D")[2].text.split("|")[0].split(" ")[0]
        except:
            None
        try:    
            Front_Camera = sp2.find_all("li",class_="rgWa7D")[2].text.split("|")[1].split(" ")[1]
        except:
            None
        try: 
            Battery = sp2.find_all("li",class_="rgWa7D")[3].text[:4]
        except:
            None
        try:    
            Processer = sp2.find_all("li",class_="rgWa7D")[4].text.split(" ")[0]
        except:
            None
        try:    
            Price = sp2.find_all("div",class_="_30jeq3 _1_WHN1")[0].text[1:]
        except:
            None
        try:    
            Link = "https://www.flipkart.com" + sp2.find("a",class_="_1fQZEK").get("href")
        except:
            None


        Multiple_Pages.append([Model_Name,Rating,Ram,ROM,Screen_Size,Rear_Camera,Front_Camera,Battery,Processer,Price,Link])    
```

```python
Multiple_Pages_Data = pd.DataFrame(Multiple_Pages,columns=["Model_Name","Rating","Ram(GB)","ROM(GB)","Screen_Size(in CM)","Rear_Camera(MP)","Front_Camera(MP)","Battery(MaH)","Processer","Price","Link_Of_Product"])
Multiple_Pages_Data
```

```python
Multiple_Pages_Data.to_csv("Final_Data.csv",index=False)
```
