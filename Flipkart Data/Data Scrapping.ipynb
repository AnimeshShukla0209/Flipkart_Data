```python
# Importing Important Libraries
from tqdm import tqdm
import pandas as pd
import numpy as np
from tqdm import tqdm
import requests
import time

from bs4 import BeautifulSoup
from selenium import webdriver
```

```python
import time
from tqdm import tqdm

options = webdriver.EdgeOptions()
driver = webdriver.Edge(options=options)
driver.get("https://www.flipkart.com/search?q=phones&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off")
```

```python
soup = BeautifulSoup(driver.page_source, "html.parser")  
```

```python
First_Page = []
for sp2 in (soup.find_all("div",class_="_2kHMtA")):
    try:
        Model_Name = sp2.find_all("div",class_="_4rR01T")[0].text.split("(")[0][:-1]
    except:
        None
    try:
        Rating = sp2.find_all("span",class_="_1lRcqv")[0].text
    except:
        None
    try:    
        Ram = sp2.find_all("li",class_="rgWa7D")[0].text.split("|")[0][:-7]
    except:
        None
    try:
        ROM = sp2.find_all("li",class_="rgWa7D")[0].text.split("|")[1][:-7]
    except:    
        None
    try:    
        Screen_Size = sp2.find_all("li",class_="rgWa7D")[1].text[:5] + " cm"
    except:    
        None
    try:    
        Rear_Camera = sp2.find_all("li",class_="rgWa7D")[2].text.split("|")[0].split(" ")[0]
    except:
        None
    try:    
        Front_Camera = sp2.find_all("li",class_="rgWa7D")[2].text.split("|")[1].split(" ")[1]
    except:
        None
    try: 
        Battery = sp2.find_all("li",class_="rgWa7D")[3].text[:4]
    except:
        None
    try:    
        Processer = sp2.find_all("li",class_="rgWa7D")[4].text.split(" ")[0]
    except:
        None
    try:    
        Price = sp2.find_all("div",class_="_30jeq3 _1_WHN1")[0].text[1:]
    except:
        None
    try:    
        Link = "https://www.flipkart.com" + sp2.find("a",class_="_1fQZEK").get("href")
    except:
        None
        
        
    First_Page.append([Model_Name,Rating,Ram,ROM,Screen_Size,Rear_Camera,Front_Camera,Battery,Processer,Price,Link])    
```

```python
First_Page_df = pd.DataFrame(First_Page,columns=["Model_Name","Rating","Ram","ROM","Screen_Size","Rear_Camera","Front_Camera","Battery","Processer","Price","Link"])
First_Page_df
```

```python
First_Page_df.to_csv("First_Data.csv",index=False)
```
